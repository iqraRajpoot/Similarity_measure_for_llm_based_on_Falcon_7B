```python
import torch
import torch.nn.functional as F
import gradio as gr
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean
import numpy as np
import re
from typing import Dict, List, Tuple
class MentalHealthValidator:
    def __init__(self):
        self.mental_health_keywords = {
            'general': [
                'anxiety', 'depression', 'stress', 'mental health', 'therapy', 'counseling',
                'psychiatrist', 'psychologist', 'mood', 'emotion', 'feeling', 'worried',
                'sad', 'anxious', 'panic', 'trauma', 'crisis', 'support', 'help',
                'mental', 'psychological', 'emotional', 'counsellor', 'treatment',
                'coping', 'struggle', 'difficulty', 'challenging', 'overwhelmed'
            ],
            'self_worth': [
                'self-image', 'self-esteem', 'self-worth', 'confidence', 'self-doubt',
                'screw-up', 'failure', 'failing', 'not enough', 'inadequate', 'unworthy',
                'worthless', 'useless', 'disappointing', 'disappointment', 'shame',
                'ashamed', 'embarrassed', 'insecure', 'inferior', 'flawed', 'broken',
                'mess up', 'messed up', 'screw up', 'screwed up', 'let down',
                'not good enough', 'self-conscious', 'self-criticism', 'self-hate',
                'self-loathing', 'self-judgment', 'judge myself', 'judging myself'
            ],
            'body_image': [
                'overweight', 'fat', 'ugly', 'unattractive', 'body image', 'look bad',
                'weight', 'appearance', 'physical appearance', 'body shape', 'body size',
                'too big', 'too small', 'not attractive', 'looks', 'looking', 'mirror'
            ],
            'emotional_states': [
                'feeling down', 'feeling sad', 'feeling anxious', 'feeling scared',
                'feeling lost', 'feeling alone', 'feeling isolated', 'feeling trapped',
                'feeling stuck', 'feeling worthless', 'feeling hopeless', 'feeling empty',
                'feeling numb', 'feeling tired', 'feeling exhausted', 'feeling overwhelmed',
                'feeling frustrated', 'feeling angry', 'feeling guilty', 'feeling ashamed',
                'feeling inadequate', 'feeling insecure', 'feeling rejected', 'feeling hurt'
            ],
            'behavioral_patterns': [
                'hiding', 'avoid', 'avoiding', 'isolation', 'withdrawn', 'withdrawing',
                'shutting down', 'shut down', 'closing off', 'closed off', 'pushing away',
                'push away', 'running away', 'run away', 'escape', 'escaping',
                'not socializing', 'not going out', 'staying in', 'stay away',
                'distance myself', 'distancing', 'isolating myself'
            ],
            'relationship_concerns': [
                'relationship', 'family', 'friend', 'friendship', 'partner', 'spouse',
                'marriage', 'divorce', 'breakup', 'separation', 'conflict', 'fight',
                'arguing', 'argument', 'disagreement', 'misunderstanding', 'trust',
                'betrayal', 'loyalty', 'commitment', 'connection', 'disconnect',
                'disconnected', 'distant', 'close', 'intimate', 'intimacy'
            ],
            'social_anxiety': [
                'social anxiety', 'social fear', 'fear of people', 'fear of judgment',
                'fear of rejection', 'fear of criticism', 'social situation',
                'social gathering', 'public speaking', 'meeting people', 'new people',
                'social interaction', 'social pressure', 'peer pressure'
            ],
            'improvement_seeking': [
                'improve', 'improvement', 'better', 'change', 'changing', 'steps',
                'help', 'advice', 'guidance', 'support', 'direction', 'path',
                'journey', 'progress', 'growth', 'growing', 'develop', 'development',
                'work on myself', 'self-improvement', 'self-help', 'self-development'
            ],
            'cognitive_patterns': [
                'thoughts', 'thinking', 'mind', 'believe', 'believing', 'thought pattern',
                'negative thoughts', 'positive thoughts', 'mindset', 'attitude',
                'perspective', 'view', 'outlook', 'perception', 'mental state',
                'mental space', 'headspace', 'mind state', 'consciousness'
            ],
            'expressions_of_distress': [
                'cant take it', "can't take it", 'too much', 'at my limit',
                'breaking point', 'rock bottom', 'end of my rope', 'cant handle',
                "can't handle", 'cant cope', "can't cope", 'cant deal',
                "can't deal", 'cant stand', "can't stand", 'cant bear',
                "can't bear", 'struggling with', 'dealing with', 'coping with'
            ]
        }

        all_keywords = []
        for category in self.mental_health_keywords.values():
            all_keywords.extend(category)
        self.pattern = re.compile(r'\b(' + '|'.join(map(re.escape, all_keywords)) + r')\b', re.IGNORECASE)

        self.emotional_phrases = [
            r"(?:I'?m|I\s+am)\s+(?:not|too|so|very|really|just|kind\s+of|sort\s+of)\s+(?:\w+)",
            r"(?:feel|feeling)\s+like\s+(?:I|me|my)",
            r"(?:don'?t|do\s+not)\s+(?:know|understand|get|see|feel)",
            r"(?:trying|tried|try)\s+(?:to|not\s+to)",
            r"(?:can'?t|cannot|cant)\s+(?:seem\s+to|help|stop|handle|deal|cope)",
            r"(?:need|want|wish)\s+to\s+(?:stop|change|improve|better|work\s+on)",
            r"(?:how\s+(?:do|can|should)\s+I)\s+(?:stop|start|begin|improve|change|cope|deal)",
            r"(?:always|never|constantly|forever)\s+(?:feel|feeling|trying|worried|concerned)",
            r"(?:no\s+one|nobody|nothing)\s+(?:understands|helps|cares|matters)",
            r"(?:everyone|everybody)\s+(?:else|around\s+me|I\s+know)",
            r"(?:hate|hating|dislike)\s+(?:myself|my\s+life|my\s+body|who\s+I\s+am)",
            r"(?:should|must|need\s+to|have\s+to)\s+be\s+(?:better|different|more|less)",
            r"(?:not\s+(?:good|smart|strong|brave|capable|worthy)\s+enough)"
        ]
        self.phrase_pattern = re.compile('|'.join(self.emotional_phrases), re.IGNORECASE)

    def is_mental_health_query(self, text: str, threshold: float = 0.08) -> Tuple[bool, float, List[str]]:
        """Enhanced validation with better context understanding"""
        if len(text.strip()) < 2:
            return False, 0.0, []

        keyword_matches = self.pattern.findall(text.lower())
        phrase_matches = self.phrase_pattern.findall(text.lower())

        all_matches = keyword_matches + phrase_matches
        unique_matches = list(set(all_matches))

        words = text.split()
        if not words:
            return False, 0.0, []

        keyword_ratio = len(all_matches) / len(words)
        phrase_bonus = len(phrase_matches) * 0.2
        context_length_bonus = min(0.2, len(words) / 100)

        emotional_state_bonus = 0.1 if any(word in text.lower() for word in ['feel', 'feeling', 'felt']) else 0
        question_bonus = 0.1 if '?' in text else 0
        personal_context_bonus = 0.1 if any(word in text.lower() for word in ['i', 'me', 'my', 'myself']) else 0

        confidence = min(1.0, (keyword_ratio * 2 +
                             len(unique_matches) * 0.15 +
                             phrase_bonus +
                             context_length_bonus +
                             emotional_state_bonus +
                             question_bonus +
                             personal_context_bonus))

        adjusted_threshold = max(0.05, threshold - (len(words) / 1000))
        is_valid = keyword_ratio >= adjusted_threshold or confidence > 0.3

        return is_valid, confidence, unique_matches

    def get_response_for_invalid_query(self) -> str:
        return ("I am a mental health assistant and can only help with questions related to "
                "mental health, emotional well-being, and psychological support. Please rephrase "
                "your question to focus on mental health concerns.")

class LightweightAttention(torch.nn.Module):
    def __init__(self, embed_dim):
        super().__init__()
        self.attention_weights = torch.nn.Linear(embed_dim, 1)

    def forward(self, embeddings, mask=None):
        embeddings = embeddings.to(self.attention_weights.weight.dtype)
        if mask is not None:
            mask = mask.to(embeddings.device)

        attention_scores = self.attention_weights(embeddings).squeeze(-1)
        if mask is not None:
            attention_scores = attention_scores.masked_fill(mask == 0, float('-inf'))
        attention_probs = F.softmax(attention_scores, dim=-1)
        attended_embeddings = torch.sum(embeddings * attention_probs.unsqueeze(-1), dim=1)
        return attended_embeddings

def get_embeddings(text, tokenizer, model, device):
    inputs = tokenizer(
        text,
        return_tensors='pt',
        truncation=True,
        padding=True,
        max_length=512
    )
    if 'token_type_ids' in inputs:
        del inputs['token_type_ids']

    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs, output_hidden_states=True)
        attention_mask = inputs['attention_mask']
        token_embeddings = outputs.hidden_states[-1].float()

        attention = LightweightAttention(token_embeddings.size(-1))
        attention = attention.to(device).to(token_embeddings.dtype)

        embeddings = attention(token_embeddings, attention_mask)
    return embeddings.squeeze()

def compute_similarities(text1, text2, tokenizer, model, device):
    embedding1 = get_embeddings(text1, tokenizer, model, device).float()
    embedding2 = get_embeddings(text2, tokenizer, model, device).float()

    embedding1_norm = F.normalize(embedding1, p=2, dim=-1)
    embedding2_norm = F.normalize(embedding2, p=2, dim=-1)

    emb1_np = embedding1_norm.cpu().detach().numpy()
    emb2_np = embedding2_norm.cpu().detach().numpy()

    cosine_raw = float(cosine_similarity(emb1_np.reshape(1, -1), emb2_np.reshape(1, -1))[0][0])
    euclidean_dist = float(euclidean(emb1_np, emb2_np))
    max_distance = np.sqrt(2.0)
    euclidean_raw = float(1.0 - (euclidean_dist / max_distance))

    length_factor = length_penalty(text1, text2)
    cosine_sim = semantic_scaling(cosine_raw) * length_factor
    euclidean_sim = semantic_scaling(euclidean_raw) * length_factor

    if cosine_raw >= 0.9:
        cosine_weight, euclidean_weight = 0.8, 0.2
    elif cosine_raw >= 0.7:
        cosine_weight, euclidean_weight = 0.7, 0.3
    else:
        cosine_weight, euclidean_weight = 0.6, 0.4

    metric_diff = abs(cosine_sim - euclidean_sim)
    agreement_factor = 1.0 - (metric_diff * 0.5)
    composite_raw = (cosine_weight * cosine_sim + euclidean_weight * euclidean_sim)
    composite_similarity = composite_raw * agreement_factor

    semantic_confidence = min(cosine_sim, euclidean_sim)
    metric_agreement = 1.0 - metric_diff
    confidence = semantic_confidence * metric_agreement * length_factor

    return {
        'cosine_similarity': float(cosine_sim),
        'euclidean_similarity': float(euclidean_sim),
        'composite_similarity': float(composite_similarity),
        'confidence': float(confidence)
    }

def length_penalty(text1, text2):
    len1 = len(text1.split())
    len2 = len(text2.split())
    ratio = min(len1, len2) / max(len1, len2)
    return np.power(ratio, 0.3)

def semantic_scaling(similarity):
    scaled = np.power(similarity, 1.5)
    if scaled >= 0.95:
        return 0.85 + (scaled - 0.95) * 3
    elif scaled >= 0.85:
        return 0.70 + (scaled - 0.85) * 1.5
    elif scaled >= 0.75:
        return 0.50 + (scaled - 0.75) * 2
    elif scaled >= 0.65:
        return 0.30 + (scaled - 0.65) * 2
    else:
        return max(0.1, scaled * 0.3)

def generate_text_with_reference_and_similarity(prompt, referenced_answer, max_length, temperature):
    try:
        # Initialize validator
        validator = MentalHealthValidator()
        is_valid, confidence, matches = validator.is_mental_health_query(prompt)

        # If not a valid mental health query, return early with explanation
        if not is_valid:
            return (validator.get_response_for_invalid_query(),
                    "",
                    "Not a mental health related query. Please rephrase your question.")

        # Continue with generation if query is valid
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        model.to(device)

        inputs = tokenizer(prompt, return_tensors='pt').to(device)
        if 'token_type_ids' in inputs:
            inputs.pop('token_type_ids')

        outputs = model.generate(
            **inputs,
            max_length=int(max_length),
            temperature=float(temperature),
            do_sample=True
        )
        generated_response = tokenizer.decode(outputs[0], skip_special_tokens=True)

        similarity_results = compute_similarities(prompt, referenced_answer, tokenizer, model, device)

        similarity_output = (
            # f"Query Validation:\n"
            # f"Mental Health Relevance: {confidence:.3f}\n"
            # f"Matched Keywords: {', '.join(matches)}\n\n"
            # f"Similarity Metrics:\n"
            # f"Cosine Similarity: {similarity_results['cosine_similarity']:.3f}\n"
            # f"Euclidean Similarity: {similarity_results['euclidean_similarity']:.3f}\n"
            f"Similarity Score: {similarity_results['composite_similarity']:.3f}\n"
            # f"Confidence: {similarity_results['confidence']:.3f}"
        )

        return generated_response, referenced_answer, similarity_output

    except Exception as e:
        return f"Error: {str(e)}", "", ""

```

```python
import base64
import gradio as gr

# Convert the logo image to base64 for embedding
def convert_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

# Path to the logo image
logo_path = "/content/drive/MyDrive/university_logo/university-of-huddersfield-logo-png-transparent (1).png"
logo_base64 = convert_to_base64(logo_path)

# Gradio Interface within Blocks
with gr.Blocks() as demo:
    # Add the logo above the interface
    gr.Markdown(
        f"""
        <div style="text-align: center; margin-bottom: 0px;">
            <img src='data:image/png;base64,{logo_base64}' alt='University Logo'
                 style='max-width: 30%; height: auto; display: block; margin-left: auto; margin-right: auto;' />
        </div>
        """
    )
    gr.Markdown(
        """
        <div style="text-align: center; margin-top: 0px; font-size: 18px;">
            This app is developed by <strong>Iqra Kiran</strong> for project purposes at the
            <strong>University of Huddersfield</strong> under the supervision of <strong>Dr. Tianhua Chen</strong>.
        </div>
        """
    )
    # Embed the interface directly in Blocks
    gr.Interface(
        fn=generate_text_with_reference_and_similarity,
        inputs=[
            gr.Textbox(lines=2, label="Input Prompt", placeholder="Ask your question..."),
            gr.Textbox(lines=2, label="Referenced Answer", placeholder="Enter your manual answer..."),
            gr.Slider(50, 500, step=50, value=300, label="Max Length"),
            gr.Slider(0.1, 1.0, step=0.1, value=0.7, label="Temperature"),
        ],
        outputs=[
            gr.Textbox(label="Generated Response"),
            gr.Textbox(label="Referenced Answer"),
            gr.Textbox(label="Generated Similarity Measure"),
        ],
        title="Custom Falcon-7B Mental Health Assistant",
        description="This app generates thoughtful responses, allows referencing answers, and includes a placeholder for similarity measure.",
    )

# Launch the application
demo.launch()
```
